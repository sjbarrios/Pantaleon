```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Distribuciones de Pol en Cachaza, Pol en Bagazo, Extracción de Sacarosa y Pol en Miel Final Zafra 2022-2023 Guatemala y Nicaragua.

     - Pol en Cachaza: Menos es mejor.
     - Pol en Bagazo: Menos es mejor y está relacionada con la Extracción de Sacarosa (menos pol de bagazo --> más extracción de sacarosa).
     - Extracción de Sacarosa: Más es mejor.
     - Pol Miel Final: Menos es mejor.

Se muestran las distribuciones de frecuencia de estos indicadores durante la zafra recién pasada. 

También se indican las metas comprometidas (mejora) por mejoras al proceso y por mejoras esperadas por modelos (en algunos casos estas son iguales, es decir, no se espera beneficio adicional por AI).

***Nota: Estas metas se establecieron antes de la zafra que recién termina, por lo que algunas metas no están alineadas con el desempeño actual, y es necesario reformularlas.***

### KPIs Proceso Guatemala (Ingenio Pantaleón)

```python
df = pd.read_csv(r'C:\Users\sbarrios\OneDrive - Universidad Galileo\Data Science\Python\Proyectos\Pruebas\Pantaleon\C3\Histogramas GT.csv')
df
```

```python
sns.set(style="whitegrid")  
fig,axs = plt.subplots(2,2, figsize = (12,10))

# Pol en Cachaza
sns.histplot(data=df, x="Pol en Cachaza", kde=True, color="skyblue", ax=axs[0, 0])
axs[0, 0].axvline(1.55 - 0.1, 0,1,color='g',linestyle='--')
axs[0, 0].text(1.55,10,'Expected Process & AI Goal: 1.55',rotation=90)

axs[0, 0].axvline(df['Pol en Cachaza'].mean()-0.1, 0,1, color ='b')
axs[0, 0].text(df['Pol en Cachaza'].mean(),10,'Last Season Process Mean: {:.2f}'.format(df['Pol en Cachaza'].mean()),rotation=90)

# Pol en Bagazo
sns.histplot(data=df, x="Pol en Bagazo", kde=True, color="olive", ax=axs[0, 1])

axs[0, 1].axvline(1.60 - 0.01, 0,1,color='r',linestyle='--')
axs[0, 1].text(1.60,10,'Expected Process Goal: 1.60',rotation=90)

axs[0, 1].axvline(1.53 - 0.01, 0,1,color='g',linestyle='--')
axs[0, 1].text(1.53,10,'Expected AI Goal: 1.53',rotation=90)

axs[0, 1].axvline(df['Pol en Bagazo'].mean()-0.01, 0,1, color ='b')
axs[0, 1].text(df['Pol en Bagazo'].mean(),10,'Last Season Process Mean: {:.2f}'.format(df['Pol en Bagazo'].mean()),rotation=90)

# Extracción de Sacarosa 
sns.histplot(data=df, x="Extracción Sacarosa", kde=True, color="gold", ax=axs[1, 0])

axs[1, 0].axvline(96.66 - 0.01, 0,1,color='r',linestyle='--')
axs[1, 0].text(96.66,10,'Expected Process Goal: 96.66',rotation=90)

axs[1, 0].axvline(96.81 - 0.01, 0,1,color='g',linestyle='--')
axs[1, 0].text(96.81,10,'Expected AI Goal: 96.81',rotation=90)

axs[1, 0].axvline(df['Extracción Sacarosa'].mean()-0.01, 0,1, color ='b')
axs[1, 0].text(df['Extracción Sacarosa'].mean(),10,'Last Season Process Mean: {:.2f}'.format(df['Extracción Sacarosa'].mean()),rotation=90)


# Pol de Miel Final
sns.histplot(data=df, x="Pol Miel Final", kde=True, color="teal", ax=axs[1, 1])

axs[1, 1].axvline(29.48 - 0.01, 0,1,color='r',linestyle='--')
axs[1, 1].text(29.48,10,'Expected Process Goal: 29.48',rotation=90)

axs[1, 1].axvline(28.56 - 0.01, 0,1,color='g',linestyle='--')
axs[1, 1].text(28.56,10,'Expected AI Goal: 28.56',rotation=90)


axs[1, 1].axvline(df['Pol Miel Final'].mean(), 0,1, color ='b')
axs[1, 1].text(df['Pol Miel Final'].mean()-0.25,10,'Last Season Process Mean: {:.2f}'.format(df['Pol Miel Final'].mean()),rotation=90)


fig.tight_layout()
```
