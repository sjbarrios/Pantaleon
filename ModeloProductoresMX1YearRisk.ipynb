## Asignación de Recursos Financieros Productores de Caña MX

### Descripción del Problema
Los Productores de Caña en México se han enfrentado a *dificultades de productividad*, debidas principalmente a condiciones climáticas atípicas y escasez de agua. Para mitigar el impacto de esta variabilidad se hace necesaria la asignación de recursos financieros, con el objetivo de garantizar la cantidad de caña que requiere la operación de los ingenios.

Esta asignación se plantea como un **problema de optimización de recursos financieros** (propios o de entidades) que permita asegurar la producción de caña requerida por los Ingenios, sujeta a un presupuesto finito y a la capacidad de re-pago de los productores.

Se formula un ***Problema de Optimización Lineal*** con los siguientes parámetros:

- **Elegibilidad (0-1)**: Este parámetro indicará si un Productor de Caña específico *califica* para una ayuda financiera basado en distintos criterios, tales como productividad mínima, monto máximo a financiar, interés en seguir produciendo caña, etc. Constituye un filtro inicial.
- **Beneficio (USD)**: Este parámetro estima el beneficio que PSA (o las entidades financieras) obtiene(n) al final de cada año por financiar a un productor de caña específico, que incluye los intereses que devenga el préstamo, como también los beneficios estimados relacionados a la productividad esperada de sus terrenos destinados a siembra de caña en términos de azúcar y energía para el grupo.
- **Monto a Financiar (USD)**: Magnitud de la ayuda financiera que el productor requiere.
- **Área Efectiva (ha)**: Se refiere al área efectiva de caña que se proyecta para los terrenos del productor destinados a siembra de caña.
- **Rendimiento de Caña (t Caña/ ha)**: Parámetro de productividad de caña por unidad de área efectiva.
- **Presupuesto (USD)**: Monto total anual de ayuda a asignar, durante el horizonte de tiempo total.
- **Total de Caña (t Caña)**: Cantidad total de caña anual requerida de los productores en necesidad de ayuda financiera.
- **Pérdida Financiera Esperada (USD)**: Estimación estadística del promedio de pérdidas de la cartera, en la que se esperaría incurrir en caso de incumplimiento de pago por parte de los productores.

### Estimación del Beneficio
**Los beneficios calculados se basan sobre la caña proporcionada por un Productor al que se le asigna recurso financiero**. Un productor puede o no recibir recursos financieros, pero para este análisis en particular, se contempla el beneficio que se obtiene dada la condición de que se asigna un monto financiero en condición de préstamo al productor.

Por esta razón, el beneficio se cuantifica en dos rubros:
- *Beneficio por Intereses*: Se asume una tasa de interés sobre el monto asignado, siendo la tasa asignada función de éste último.
- *Beneficio por Caña*: La caña que se obtiene que se obtiene del productor dado que le fue asignado un monto, función de la cantidad de azúcar y energía que se obtiene de la misma.

$$ Beneficio_{esperado} = Beneficio_{caña} + Beneficio_{intereses}$$

### Estimación del Riesgo Financiero
La estimación del riesgo financiero que se incurre al asignar un monto en condiciones de préstamo a un productor es una de las variables más importantes en la selección de productores. Se propone la utilización de la $Perdida_{esperada}$ como estimación de este riesgo, descrita por la ecuación:
$$Perdida_{esperada} = Exposición_{impago} * Pérdida_{impago} * Riesgo_{impago}$$

Donde:
- **Exposición al Momento de Impago (USD)**: Monto total de la ayuda asignada al productor. En este caso igual al Monto a Financiar.
- **Pérdida dado el Impago (%)**: Fracción de pérdida respecto al monto total en la que se incurriría en caso de impago. Esta fracción normalmente es menor que 1 dado que en ocasiones existen garantías sobre el préstamo.
- **Riesgo de Impago (%)**: Probabilidad de un productor de caña de no cumplir con los compromisos de pago.

De las cantidades anteriores, la más difícil de determinar es el **Riesgo de Impago (%)** (también conocido como *Probability of Default*). Normalmente es función de las condiciones económicas del país, las características socio-económicas del solicitante y otros factores externos. Las herramientas utilizadas comprenden el "credit scoring", modelos probabilísticos de estimación basados en históricos (normalmente modelos de regresión logística o de machine learning). Estos enfoques sin embargo requieren disponer de una historia crediticia del solicitante. Para el caso específico en estudio, no se cuenta con esta información por lo que se propone un enfoque alternativo.

#### Cociente Deuda a Ingresos
El cociente *Deuda a Ingreso* (Income-Debt Ratio) es el porcentaje del Ingreso Mensual Bruto que se destina a cubrir obligaciones de Deuda, entre ellas *préstamos, tarifas, impuestos, primas de seguros, etc*. Este cociente (en conjunto con el "credit scoring") es la variable más relevante en la determinación de la *Probabilidad de Impago* y es más sencillo de obtener porque requiere de información crediticia básica del productor.

Debido a esto, formulamos un modelo de *Regresión Logística* de una variable, con la forma:

$$ DefaultProbability = \frac {e^{(\beta_0 + \beta_1 * IncomeDebtRatio)}}{1+e^{(\beta_0 + \beta_1 * IncomeDebtRatio)}}$$

Este modelo supone determinar los coeficientes $\beta_0$ y $\beta_1$ óptimos para la estimación correcta del riesgo. Normalmente esto se realiza de forma similar a una regresión con datos históricos. En este caso no se cuenta con esta información, por lo que se propone un enfoque alternativo por medio de una **estimación de dos puntos**, en la cual se resuelven dos ecuaciones con dos incógnitas.

Suponga que se tiene información de los productores en la que:
- Un cociente deuda-ingreso $R_0$ se asocia con una probabilidad de incumplimiento $P_0$
- Un cociente deuda-ingreso $R_1$ se asocia con una probabilidad de incumplimiento $P_1$

Se resuelven las ecuaciones para encontrar los parámetros de la regresión, con el siguiente resultado:
$$\beta_1 = \frac{1}{R_0-R_1} * [ln(\frac{P_0}{1-P_0}) - ln(\frac{P_1}{1-P_1})]$$
$$\beta_0 = ln(\frac{P_0}{1-P_0}) - \beta_1*R_0$$

Esto permite estimaro la **Probabilidad de Impago** dados dos puntos de referencia relacionados al **Cociente Deuda-Ingreso**

### Formulación del Problema de Optimización

#### Conjuntos y Notaciones:

Sean:
- $I$ el conjunto de Productores de Caña, con $i$ el i-ésimo productor de caña, $i \in \{1,2,...,n\}$.
- $J$ el conjunto de Años en el Horizonte de Análisis, con $j$ el j-ésimo año en el horizonte, $j \in \{1,2,...,m\}$.

#### Parámetros:
- $E_{i} \in \{0,1\}, \forall i \in \{1,2,...,n\} $ la Elegibilidad a Financiamiento del i-ésimo Productor de Caña.
- $B_{i,j} \in \mathbb{R}, \forall i \in \{1,2,...,n\}, \forall j \in \{1,2,...,m\}$ el Beneficio Estimado (NPV) por Financiar al Productor $i$ en el año $j$.
- $F_{i,j} \in \mathbb{R}, \forall i \in \{1,2,...,n\}$ el Financiamiento Requerido por el i-ésimo Productor de Caña.
- $A_{i,j} \in \mathbb{R}, \forall i \in \{1,2,...,n\}, \forall j \in \{1,2,...,m\}$ el Área Efectiva de Caña del Productor $i$ en el año $j$.
- $C_{i,j} \in \mathbb{R}, \forall i \in \{1,2,...,n\}, \forall j \in \{1,2,...,m\}$ el Rendimiento Ponderado de Caña (TCH) del Productor $i$ en el año $j$.
- $P_{i,j} \in \mathbb{R}, \forall i \in \{1,2,...,n\}$, la Pérdida Esperada de cada productor $i$, por riesgo de impago en el año $j$
- $L_{i,j} \in [0,1], \forall i \in \{1,2,...,n\}$, la Pérdida Financiera dado el impago del Productor de Caña $i$ en el año $j$
- $R_{i,j} \in [0,1], \forall i \in \{1,2,...,n\}$, el Riesgo de Impago del Productor de Caña $i$ en el año $j$

#### Variables de Decisión:
- $X_i \in \{0,1\}, \forall i \in \{1,2,...,n\}$ la variable que indica la decisión de financiar al i-ésimo Productor de Caña en todo el horizonte de análisis.

#### Función Objetivo Beneficio Financiación Productores MX

$$\max\limits_{\forall i \in I, \forall j \in J} \sum_{i=1}^{n}\sum_{j=1}^{m} X_{i}*E_{i}*B_{i,j} - \sum_{i=1}^{n}\sum_{j=1}^{m} P_{i,j}*X_{i}$$

Donde:
$$ P_{i,j} = F_{i,j}*L_{i,j}*R_{i,j}, \forall i \in I, \forall j \in J $$

#### Restricciones

Presupuesto Anual para Financiamiento:

- $\sum_{i=1}^{n} X_{i}*F_{i,j} \leq Budget_{j}; \forall j \in J$

Demanda Anual de Caña:

- $\sum_{i=1}^{n} X_i*A_{i,j}*C_{i,j} + Caña_{Propia} + Caña_{Productores Seguros} \geq CañaMinima_{j};  \forall j \in J$

- $\sum_{i=1}^{n} X_i*A_{i,j}*C_{i,j} + Caña_{Propia} + Caña_{Productores Seguros} \leq CañaMaxima_{j};  \forall j \in J$

No-Negatividad del Beneficio:

- $\sum_{i=1}^{n}\sum_{j=1}^{m} X_{i}*E_{i}*B_{ij} \geq 0$

Tolerancia Máxima al Riesgo de Incumplimiento (Probabilidad de Impago):

- $R_{i,j} \leq R_{MAX}, \forall i \in \{1,2,...,n\}, \forall j \in J$

Monto de Financiamiento Máximo a Productor Individual:

- $F_{i,j} \leq F_{MAX}, \forall i \in \{1,2,...,n\}, \forall j \in \{1,2,...,m\}$

Variables de Decisión:

- $X_i \in \{0,1\}, \forall i \in \{1,2,...,n\}$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# librería de optimización
import gurobipy as grb

#from google.colab import drive
#drive.mount('/content/drive')
```

### Datos de Productores de Caña

```python
df = pd.read_csv(r'ProductoresMX_1Years.csv')
df
```

### Estimación del Beneficio

```python
# Azúcar
KARBE = 105         # kg Azúcar / t Caña
BENEFICIO_AZUCAR = 467.06  # USD / t Azúcar Cruda
PORCENTAJE_PRODUCTOR = 0.43 # % del Beneficio Total del Azúcar asignado al Productor

# Energía
EFF_EN = 89.00     # kWh / t Caña
BENEFICIO_ENERGIA = 40 # USD / MWh

# Tasa de Interés
INT_RATE = 0.1     # Tasa de Interés Anual

df['Beneficio_Azucar_Year1'] = df['Rendimiento_Year1'] * df['Area_Year1'] * KARBE/1000 * BENEFICIO_AZUCAR * (1-PORCENTAJE_PRODUCTOR)
df['Beneficio_Energia_Year1'] = df['Rendimiento_Year1'] * df['Area_Year1'] * EFF_EN/1000 * BENEFICIO_ENERGIA
df['Beneficio_Interes_Year1'] = df['Financiamiento_Year1'] * INT_RATE
df['Beneficio_Year1'] = df['Beneficio_Azucar_Year1'] + df['Beneficio_Energia_Year1'] + df['Beneficio_Interes_Year1']
df.drop(['Beneficio_Azucar_Year1','Beneficio_Energia_Year1','Beneficio_Interes_Year1'], axis='columns', inplace=True)
df
```

### Estimación de Riesgo Financiero

```python
# Estimación de Dos Puntos (Income_Debt_Ratio, Probability_of_Default)
R_0 = 0.3
P_0 = 0.2

R_1 = 0.5
P_1 = 0.8
```

```python
# Coeficientes de Regresión Logística
beta_1 = (1/(R_0-R_1))*(np.log(P_0/(1-P_0))-np.log(P_1/(1-P_1)))
beta_0 = np.log(P_0/(1-P_0))-beta_1*R_0

# Curva Probabilidades vs. Ratios
ratios = np.arange(0,1.01,0.01)
probabilities = 1/(1+np.exp(-(beta_0+beta_1*ratios)))
d = {"income_debt_ratio":ratios,"probability_of_default":probabilities}
prob_df = pd.DataFrame(d)
sns.set_style("darkgrid")
sns.scatterplot(x="income_debt_ratio",y="probability_of_default",data=prob_df).set(title='Probabilidad de Incumplimiento Financiero (%)')
```

```python
df['Income_Debt_Ratio_Year1'] = df['Deudas_Year1'] / df['Ingresos_Year1']
df['Probabilidad_Impago_Year1'] = 1/(1+np.exp(-(beta_0+beta_1*df['Income_Debt_Ratio_Year1'])))
df.drop(['Ingresos_Year1','Deudas_Year1','Income_Debt_Ratio_Year1'], axis='columns', inplace=True)
df
```

#### Parámetros de Optimización

```python
## productores
Producers = np.array(df)

# número de productores
n = len(df['ID'])

# número de años
a = 1

# lista de productores
I = {i for i in range(1,n+1)}

# lista de años
J = {j for j in range(1,a+1)}

# parámetros
BUDGET = [1000000]
CANA_MIN = [300000]
CANA_MAX = [500000]
RISK = 0.2
FUNDING = 500000

E = {i:Producers[i-1,1] for i in I} # Elegibilidad
B = {(i,j):Producers[i-1,6*(j-1)+2] for i in I for j in J} # Beneficio
F = {(i,j):Producers[i-1,6*(j-1)+3] for i in I for j in J} # Financiamiento
A = {(i,j):Producers[i-1,6*(j-1)+4] for i in I for j in J} # Área
C = {(i,j):Producers[i-1,6*(j-1)+5] for i in I for j in J} # Rendimiento
L = {(i,j):Producers[i-1,6*(j-1)+6] for i in I for j in J} # 
R = {(i,j):Producers[i-1,6*(j-1)+7] for i in I for j in J}
```

#### Modelo de Optimización

```python
# modelo de optimización
model = grb.Model("Seleccion de Productores")

# variables de decisión
x = model.addVars(I, vtype = grb.GRB.BINARY)
```

#### Función Objetivo

```python
# función objetivo
model.setObjective(grb.quicksum(E[i]*B[(i,j)]*x[i] - F[(i,j)]*L[(i,j)]*R[(i,j)]*x[i] for i in I for j in J), grb.GRB.MAXIMIZE)
```

#### Restricciones

```python
# restricciones
model.addConstrs((grb.quicksum(F[(i,j)]*x[i] for i in I) <= BUDGET[j-1] for j in J), name = "Financiamiento Anual Maximo")
model.addConstrs((grb.quicksum(C[(i,j)]*A[(i,j)]*x[i] for i in I) >= CANA_MIN[j-1] for j in J), name = "Demanda Minima Anual de Caña")
model.addConstrs((grb.quicksum(C[(i,j)]*A[(i,j)]*x[i] for i in I) <= CANA_MAX[j-1] for j in J), name = "Demanda Maxima Anual de Caña")
model.addConstr((grb.quicksum(E[i]*B[(i,j)]*x[i] for i in I for j in J) >= 0), name = "No-Negatividad del Beneficio Financiero")
model.addConstrs((R[(i,j)]*x[i] <= RISK for i in I for j in J), name = "Tolerancia al Riesgo de Incumplimiento")
model.addConstrs((F[(i,j)]*x[i] <= FUNDING for i in I for j in J), name = "Financiamiento Individual Maximo")

model.update()
```

### Solución al Problema de Optimización de Programación Entera-Mixta (MIP)

```python
# optimización
model.optimize()

if model.status != grb.GRB.status.OPTIMAL:
    print("No se encontró una solución factible...")
else:
    print("El valor optimo de la función objetivo es %g"%model.objVal)
    resultado_objetivo = model.objVal
    resultado_seleccion_productores = [i for i in I if x[i].X >= 1]

    df_salida = pd.DataFrame({"valor función objetivo": [resultado_objetivo],
                              "productores seleccionados": [resultado_seleccion_productores]})
```

```python
df_salida
```

#### Análisis de Resultados

```python
productores = df_salida['productores seleccionados'][0]
df_final = df.loc[df['ID'].isin(productores)].reset_index()
df_final.drop('index', axis=1, inplace=True)
df_final.head()
```

```python
df_final['Perdida_Esperada_Year1'] = df_final['Financiamiento_Year1']*df_final['Perdida_Impago_Year1']*df_final['Probabilidad_Impago_Year1']

fig, ax = plt.subplots(2, 3, figsize=(20, 10))

df_final['Beneficio_Year1'].hist(ax=ax[0,0],alpha=0.5,color='lightgreen')
ax[0,0].set_title('Expected Benefit Year 1 (USD)')

df_final['Perdida_Esperada_Year1'].hist(ax=ax[0,1],alpha=0.5,color='red')
ax[0,1].set_title('Expected Loss Year 1 (USD)')

df_final['Probabilidad_Impago_Year1'].hist(ax=ax[0,2],alpha=0.5,color='orange')
ax[0,2].set_title('Default Risk Year 1 (%)')

df_final['Rendimiento_Year1'].hist(ax=ax[1,0],alpha=0.5,color='darkgreen')
ax[1,0].set_title('Expected Yield Year 1 (TCH)')

df_final['Area_Year1'].hist(ax=ax[1,1],alpha=0.5,color='brown')
ax[1,1].set_title('Expected Area Year 1 (ha)')

df_final['Financiamiento_Year1'].hist(ax=ax[1,2],alpha=0.5,color='blue')
ax[1,2].set_title('Financing Year 1 (USD)')
```
